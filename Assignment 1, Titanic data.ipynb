```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('titanic_train.csv')
data.head()
```

```python
data.isnull()
```

```python
fig = plt.subplots(figsize=(12,7))
sns.set(font_scale=1.2)
sns.heatmap(data.isnull(),yticklabels=False,cbar_kws={'label': 'Color bar'},cmap='summer')
```

```python
fig = plt.subplots(figsize=(12,7))
sns.set_style('whitegrid',{'grid.color': ".5", 'grid.linestyle': ":"})
sns.countplot(x='Survived', data=data, palette='copper',saturation=1)
```

```python
fig = plt.subplots(figsize=(12,7))
sns.set_style('ticks',{'grid.color': ".5",})
sns.countplot(x='Survived',hue='Gender',data=data,palette='cubehelix',saturation=1)
```

```python
fig = plt.subplots(figsize=(12,7))
sns.set_style('darkgrid')
sns.countplot(x='Survived',hue='Pclass',data=data,palette='icefire')
```

```python
fig = plt.subplots(figsize=(12,7))
sns.distplot(data['Age'].dropna(),kde=False,rug=True,color='darkred',bins=60,hist_kws={"histtype": "step", "linewidth": 4,
                            "alpha": 1, "color": "k"})
sns.set(font_scale = 1.2)
```

```python
fig = plt.subplots(figsize=(12,7))
data['Age'].hist(bins=30,color='green',alpha=1)
```

```python
fig = plt.subplots(figsize=(12,7))
sns.countplot(x='SibSp',data=data,facecolor=(0, 0, 0, 0),linewidth=5,
                   edgecolor=sns.color_palette("dark", 3))
```

```python

sns.jointplot(kind='scatter',color='r',x="Age", y="Fare",height=10, data=data)
```

```python
data['Fare'].hist(color='cyan',bins=40,figsize=(12,7),alpha=1)
plt.xlabel('Fare')
```
